```python
import string

def freq_count():
    with open(r"/Users/piyushkumar/Desktop/myOS/MY_FILE.txt", mode="r") as file:
        text = file.read()

    pattern = str.maketrans("", "", string.punctuation)
    clean_text = text.translate(pattern)
    words_list = clean_text.lower().split()

    words_count_dict = {}

    for word in words_list:
        if word not in words_count_dict:
            words_count_dict[word] = 1
        else:
            words_count_dict[word] += 1

    for key, value in words_count_dict.items():
        print(key, value)

if __name__=="__main__":
    try:
        print(freq_count())
    except Exception as e:
        print(e)
```
