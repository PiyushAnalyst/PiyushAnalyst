```python
def generate_even_number(n: int) -> int:
    """fucntion
    """
    if not isinstance(n,int):
        raise TypeError("input must be integer")
    if n<=2:
        raise ValueError("input must be greater than or eaual to 2")
    even_num_list1 = [i for i in range(2,n+1,2)]
    return(even_num_list1)
#main exceution 
if __name__== "__main__":
    n = 100
    try:
        result = generate_even_number(n)
        print("even number list:",result)
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
    
```
