```python
def find_common_elements(list1: list, list2: list, list3: list) -> list:
    if not isinstance(list1,list)or not isinstance(list2,list)or not isinstance(list3,list):
        raise TypeError("All elements must be list")
    
    common = []
    for item in list1:
        if item in list2 and item in list3 and item not in common:
            common.append(item)

    return common

if __name__ == "__main__":
    try:
        list1 = [1, 2, 3, 4, 5]
        list2 = [3, 4, 5, 6, 7]
        list3 = [2, 3, 5, 8, 9]
        print(find_common_elements(list1,list2,list3))
    except Exception as e:
        print(e)
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
list3 = [2, 3, 5, 8, 9]

common = []

for i in list1:
    for j in list2:
        for k in list3:
            if i == j and j == k:
                if i not in common:
                    common.append(i)

print("Common elements:", common)
```
