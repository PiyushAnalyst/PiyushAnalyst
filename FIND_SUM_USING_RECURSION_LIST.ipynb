```python
list1 = [10, 20, 30, 40]

def recursive_sum(list1):
    if len(list1) == 0:
        return 0
    return list1[0] + recursive_sum(list1[1:])

print(recursive_sum(list1))
```

```python
def recursive_sum(listi: list) -> int:
    if not isinstance(listi, list):
        raise TypeError("Input must be a list")
    if len(listi) == 0:
        raise ValueError("List cannot be empty")
    # if not all(isinstance(x, (int, float)) for x in listi):
    #     raise TypeError("List can contain only numeric values")
    if len(listi) == 1:
        return listi[0]
    return listi[0] + recursive_sum(listi[1:])

# Main program
if __name__=="__main__":
    list1 = [12, 12]

    try:
        print(recursive_sum(list1))
    except (TypeError, ValueError) as e:
        print(e)
```
