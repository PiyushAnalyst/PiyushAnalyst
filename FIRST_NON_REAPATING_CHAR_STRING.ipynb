```python
text = "swwissn"
char_count = {}
for char in text:
    if char in char_count:
        char_count[char]+=1
    else:
        char_count[char] = 1
    print(char_count)
for char in text:
    if char_count[char]==1:
        print(char)
        break
```

```python
def first_non_repeated_char(text:str)->str:
    char_count = {}
    for char in text:
        if char in char_count:
            char_count[char]+=1
        else:
            char_count[char] = 1
        print(char_count)
    for char in text:
        if char_count[char]==1:
            return char
if __name__=="__main__":
    try:
        text = ""
        print(first_non_repeated_char(text))
    except Exception as e:
        print(e)
```

```python
first_non_repeated_char("abca")
```
