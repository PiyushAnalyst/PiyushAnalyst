```python
list1 = [12,13,45,67,89,90]
for i in range(len(list1)-1):
    if list1[i] > list1[i+1]:
        print(False)
        break
else:
    print(True)
```

```python
def is_sorted (list1:list)-> bool:
    for i in range(len(list1)-1):
        if not isinstance(list1[i],int) or not isinstance (list1[i +1 ],int):
            raise TypeError
        if list1[i] > list1[i+1]:
            return False


    return True


if __name__=="__main__":
    list1 = [12,13,113,31,141,31,3,1,3]
    try:
        print(is_sorted(list1))
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)

```
