```python
def find_largest(numbers: list[float]) -> float:

    # Check input is list
    if not isinstance(numbers, list):
        raise TypeError("Input must be a list")

    # Check list is not empty
    if not numbers:
        raise ValueError("Input cannot be empty")

    # Check all elements are int or float
    for item in numbers:
        if not isinstance(item, (int, float)):
            raise ValueError("All elements must be int or float")

    #  Assume first element is largest
    largest = numbers[0]

    #  Loop to find largest
    for item in numbers:
        if item > largest:
            largest = item

    return largest


if __name__ == "__main__":
    try:
        user_input = input("Enter your elements: ")

        # Convert input into list of floats
        numbers_list = [float(item) for item in user_input.split()]

        result = find_largest(numbers_list)

        print("Largest number is:", result)

    except (TypeError, ValueError) as e:
        print("Error:", e)

    except Exception as e: 
        print("Unexpected Error:", e)
```
