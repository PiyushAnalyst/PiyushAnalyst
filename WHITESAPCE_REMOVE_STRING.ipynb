```python
def whitespace(text:str)->str:
    """docstring"""
    if not isinstance(text,str):
        raise TypeError
    result = ""
    prev_space = False
    text.strip()
    for char in text.strip():
        if not char.isspace():
            result  = result + char
            prev_space = False
        else:
            if prev_space == False:
                result = result + " "
                prev_space = True
    return result
if __name__=="__main__":
    text = " \n       this  \t    is \r     python  " 
    try:
        print(whitespace(text))
    except TypeError as e:
        print(e)
        
```
