```python
# Q1.........
# Accept number of units from user
unit = int(input("Enter electricity units: "))

bill = 0

# First 100 units - no charge
if unit > 100:
    bill = bill + (100 * 0)

# Next 100 units (101–200) Rs 5 per unit
if unit > 100:
    if unit <= 200:
        bill = bill + ((unit - 100) * 5)

# After 200 units Rs 10 per unit
if unit > 200:
    bill = bill + (100 * 5)
    bill = bill + ((unit - 200) * 10)

print("Total Electricity Bill = Rs", bill)
```

```python
# Q2.........
# Accept percentage from user
percentage = float(input("Enter your percentage: "))

# Check grade
if percentage > 90:
    print("Grade A")
elif percentage > 80 and percentage <= 90:
    print("Grade B")
elif percentage >= 60 and percentage <= 80:
    print("Grade C")
else:
    print("Grade D")
```

```python
# Q3..........
# Accept age of 4 people
age1 = int(input("Enter age of person 1: "))
age2 = int(input("Enter age of person 2: "))
age3 = int(input("Enter age of person 3: "))
age4 = int(input("Enter age of person 4: "))

# Assume first age is youngest
youngest = age1

if age2 < youngest:
    youngest = age2

if age3 < youngest:
    youngest = age3

if age4 < youngest:
    youngest = age4
 

print("The youngest age is:", youngest)
```

```python
# Q4...........
# Accept salary and years of service from user
salary = float(input("Enter your salary: "))
service = int(input("Enter years of service: "))

# Calculate bonus
if service > 10:
    bonus = salary * 0.10
elif service >= 6 and service <= 10:
    bonus = salary * 0.08
else:
    bonus = salary * 0.05

print("Bonus amount is:", bonus)
```

```python
# Q5.........
# Accept three numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Find second largest
if (a > b and a < c) or (a < b and a > c):
    second = a
elif (b > a and b < c) or (b < a and b > c):
    second = b
else:
    second = c

print("Second largest number is:", second)
```

```python
# Q6.........
# Accept marked price from user
price = float(input("Enter Marked Price: "))

# Calculate discount
if price > 10000:
    discount = price * 0.20
elif price > 7000 and price <= 10000:
    discount = price * 0.15
else:
    discount = price * 0.10

# Net amount
net_amount = price - discount

print("Discount:", discount)
print("Net amount to pay:", net_amount)
```
